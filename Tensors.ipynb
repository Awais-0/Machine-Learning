<h3>Tensors</h3>
<p>The basic data structure in machine learning</p>
<p>Can be of many dimensions</p>

```python
import numpy as np
```

<p>1. 0D tensors</p>
<small>A single number having no dimensions. Just a scalar number</small>

```python
a = np.array(4)
a
```

<small>To get the dimensions of a tensor we use numpy functions like ndim, rank, shape</small>

```python
print(np.ndim(a))
print(np.shape(a))
```

<p>2. 1D Tensors</p>
<small>Collection of 0D tensors(scalars). These are same like array in programming.</br> Any Tensor having 1 or more dimensions is a vector</small>

```python
arr = np.array([1, 2, 3])
print(type(arr)) # n(1)-dimensional array
print(arr.ndim) # 1 dimension
print(arr.shape) # 3 elements on single dim(axis)
```
